```python
df['Returns'] = df['Returns'].fillna(df['Returns'].mean())
```

Checking depulicate

```python
duplicate_count = df.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicate_count}")
```

```python
# Drop the duplicate "Age" column
df.drop(columns=["Age"], inplace=True)

```

```python
# Check and remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Display cleaned dataset info
df.info()
df.head()
```

```python
# Convert "Purchase Date" to datetime format
df["Purchase Date"] = pd.to_datetime(df["Purchase Date"], errors="coerce")
```

Check for Logical Inconsistencies

```python
categorical_columns = df.select_dtypes(include=['object']).columns
print("\nUnique Values in Categorical Columns:")
for col in categorical_columns:
    print(f"{col}: {df[col].unique()[:10]}")
```

```python
# Check for outliers using IQR for numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns
print("\nOutlier Counts per Numerical Column:")
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outlier_count = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))].shape[0]
    print(f"{col}: {outlier_count} outliers")
```
